### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".
grouped_df = df.groupby(['region', 'state', 'city'])['units_sold'].sum().reset_index()
grouped_df.rename(columns={'units_sold': 'Total_Products_Sold'}, inplace=True)
grouped_df = grouped_df.sort_values('Total_Products_Sold', ascending=False).set_index(['region','state','city'])

# Show the top 5 results.
grouped_df.head()
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.
units_sold_pivot = df.pivot_table(index=['region', 'state', 'city'], values='units_sold', aggfunc='sum')

# Rename the "units_sold" column to "Total_Products_Sold"
units_sold_pivot.rename(columns={'units_sold': 'Total_Products_Sold'}, inplace=True)
units_sold_pivot = units_sold_pivot.sort_values('Total_Products_Sold', ascending=False)

# Show the top 5 results.
units_sold_pivot.head()
```

### 6. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.
# STEP 1 ####################################

invoice_pivot = womens_df.pivot_table(index='invoice_date', values='total_sales', aggfunc='sum')
invoice_pivot.head(10)
```

```python
# Optional: Rename the "total_sales" column to "Total Sales"
# STEP 2 ####################################

invoice_pivot.rename(columns={'total_sales':'Total Sales'}, inplace=True)
invoice_pivot.head(10)
```

```python
# Show the table.
# STEP 3 ####################################

invoice_pivot.head(10)
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.
# STEP 1 ####################################

invoice_pivot_daily = invoice_pivot.resample('D').sum()
invoice_pivot_daily.head(10)
```

```python
# Sort the resampled pivot table in descending order on "Total Sales".
# STEP 2 ####################################

invoice_pivot_daily.sort_values('Total Sales', ascending=False, inplace=True)
invoice_pivot_daily.head(10)
```

```python
# STEP 3 ####################################

invoice_pivot_daily.head(10)
```

### The week of 2021-12-19 had the highest sales for Women's Athletic Footwear as highlighted in the calendar below.

```python
class HighlightCalendar(calendar.TextCalendar):
    def __init__(self, firstweekday=6):
        super().__init__(firstweekday)

    def formatweek(self, theweek, highlight_week=None):
        week_str = ''
        for day, weekday in theweek:
            if highlight_week and (day, weekday) in highlight_week:
                day_str = f'[{day:2}]' if day != 0 else '    '
            else:
                day_str = f' {day:2} ' if day != 0 else '    '
            week_str += day_str
        return week_str.rstrip()
    
    def formatmonth(self, theyear, themonth, highlight_week=None):
        weeks = self.monthdays2calendar(theyear, themonth)
        formatted_month = self.formatmonthname(theyear, themonth, 0) + '\n'
        formatted_month += self.formatweekheader(3) + '\n'
        for week in weeks:
            formatted_month += self.formatweek(week, highlight_week) + '\n'
        return formatted_month.strip()

year = 2021
month = 12
week_number = 3
highlight_cal = HighlightCalendar(firstweekday=6)
weeks = highlight_cal.monthdays2calendar(year, month)
if 0 <= week_number < len(weeks):
    highlight_week = weeks[week_number]
else:
    highlight_week = None

print(highlight_cal.formatmonth(year, month, highlight_week))
```
